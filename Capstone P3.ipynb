Problem Background

When moving to a new city, there are many things to consider that can impact the decision of where to live.  One possible factor that can influence a decision is how well the city compares when faced with a zombie apocalypse.  With the rising trend in science fiction and the increased knowledge in science itself, it is only a matter of time until a zombie breakout occurs.  Since this is the inevitable outcome, deciding what city to live in can be crutial to survival.  When it comes to disaster scenarios, many people have leanings into three categories.  First group are the survivalists who prefer making a bunker to live out their days.  Second are the runners, those who hop from place to place when an area gets too crazy. Third, those that accept the doom and want a spectacular end.  In order to find the best city for the three differnt types, statistical analysis is needed on different cities to highlight the key features of each group. 

Data

In order to determine what city to live in, the five largest cities in the USA will be analyzed for different key features that will be important for each of the different group types.  The venue information will be used for each city and total counts of different categories will be grouped and analyzed.  For the three type groups, the main features are outlined below.
1. Survivalists
    Pharmacy

    Outdoor Recreation
2. Runners

3. Spectacular

Each category will receive a weight based on number of venues and proximity.  Once all the data is compiled and analyzed, there should be clear choices for each category and middle ground for those whose beliefs are in multiple categories.
